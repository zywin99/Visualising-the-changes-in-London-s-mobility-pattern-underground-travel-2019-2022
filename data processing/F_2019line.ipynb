```python
from scipy.stats import skew, kurtosis
from scipy.signal import find_peaks
columns_to_calculate = df_exit.columns[1:] 

result = pd.DataFrame()
result['Station'] = df_exit['Station']


result['Max_Flow'] = df_exit[columns_to_calculate].max(axis=1)
result['Max_Flow_Time'] = df_exit[columns_to_calculate].idxmax(axis=1)


result['Skewness'] = df_exit[columns_to_calculate].apply(lambda row: skew(row), axis=1)
result['Kurtosis'] = df_exit[columns_to_calculate].apply(lambda row: kurtosis(row), axis=1)

#  (p)
total_flow = df_exit[columns_to_calculate].sum(axis=1)
result['Peak_Hour_Coefficient'] = result['Max_Flow'] / total_flow

peak_hours = ['07:00-08:00', '08:00-09:00', '09:00-10:00', '16:00-17:00', '17:00-18:00', '18:00-19:00']
off_peak_hours = ['10:00-11:00', '11:00-12:00', '12:00-13:00', '13:00-14:00', '14:00-15:00', '15:00-16:00']

# (U)
G = df_exit[peak_hours].mean(axis=1)
H = df_exit[off_peak_hours].mean(axis=1)
result['Balance_Coefficient'] = G / H


def count_peaks(row):
    peaks, _ = find_peaks(row)
    return len(peaks)

result['Peak_Count'] = df_exit[columns_to_calculate].apply(count_peaks, axis=1)
```

```python

from geopy.distance import geodesic
from scipy.spatial import KDTree
np.random.seed(42)

# R^n 
X = merged_df.drop(columns=['Station']).values


K = 7  
m, n = X.shape
C = []  
X_c = X.copy()  
indices = np.arange(m)

# Step 1
first_center_idx = np.random.choice(indices)
C.append(X[first_center_idx])
X_c = np.delete(X_c, first_center_idx, axis=0)
indices = np.delete(indices, first_center_idx)

def calculate_distance(x, centers):
    return np.min([np.linalg.norm(x - c) for c in centers])

# Step 2-5
while len(C) < K:
    D = np.array([calculate_distance(x, C) for x in X_c])
    D_squared = D ** 2
    p = D_squared / np.sum(D_squared)
    next_center_idx = np.random.choice(indices, p=p)
    C.append(X[next_center_idx])
    X_c = np.delete(X_c, next_center_idx, axis=0)
    indices = np.delete(indices, next_center_idx)

C = np.array(C)

# Step 6-7
def assign_clusters(X, centers):
    clusters = {}
    for i, x in enumerate(X):
        min_dist = float('inf')
        assigned_center = None
        for j, c in enumerate(centers):
            dist = np.linalg.norm(x - c)
            if dist < min_dist:
                min_dist = dist
                assigned_center = j
        if assigned_center not in clusters:
            clusters[assigned_center] = []
        clusters[assigned_center].append(i)
    return clusters

clusters = assign_clusters(X, C)
prev_clusters = None

while clusters != prev_clusters:
    prev_clusters = clusters
    new_centers = []
    for k in clusters.keys():
        cluster_points = X[clusters[k]]
        new_center = np.mean(cluster_points, axis=0)
        new_centers.append(new_center)
    C = np.array(new_centers)
    clusters = assign_clusters(X, C)


cluster_labels = np.zeros(m, dtype=int)
for k, indices in clusters.items():
    cluster_labels[indices] = k

merged_df['Cluster'] = cluster_labels


print(merged_df)


#merged_df.to_csv('/Users/zoyouwe/Desktop/clustered_stations.csv', index=False)
```
