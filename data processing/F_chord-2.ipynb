```python
mnlc_to_cluster = dict(zip(dfCLU_merged['MNLC'], dfCLU_merged['Cluster']))


df_combined['mnlc_o'] = df_combined['mnlc_o'].map(mnlc_to_cluster)
df_combined['mnlc_d'] = df_combined['mnlc_d'].map(mnlc_to_cluster)


print("Updated df_combined DataFrame:")
print(df_combined)
```

```python
import json
nodes = list(set(df_combined['mnlc_o']).union(set(df_combined['mnlc_d'])))
nodes.sort()


node_indices = {node: i for i, node in enumerate(nodes)}


edges_am_peak = []
edges_inter_peak = []
edges_pm_peak = []

for _, row in df_combined.iterrows():
    source = node_indices[row['mnlc_o']]
    target = node_indices[row['mnlc_d']]
    edges_am_peak.append({"source": source, "target": target, "value": row['AM Peak']})
    edges_inter_peak.append({"source": source, "target": target, "value": row['Inter Peak']})
    edges_pm_peak.append({"source": source, "target": target, "value": row['PM Peak']})


node_data = [{"name": str(node)} for node in nodes]


def create_matrix(edges, node_count):
    matrix = [[0] * node_count for _ in range(node_count)]
    for edge in edges:
        source = edge['source']
        target = edge['target']
        value = edge['value']
        matrix[source][target] = value
    return matrix

matrix_am_peak = create_matrix(edges_am_peak, len(nodes))
matrix_inter_peak = create_matrix(edges_inter_peak, len(nodes))
matrix_pm_peak = create_matrix(edges_pm_peak, len(nodes))

data = {
    "nodes": node_data,
    "matrix_am_peak": matrix_am_peak,
    "matrix_inter_peak": matrix_inter_peak,
    "matrix_pm_peak": matrix_pm_peak
}

with open('/Users/zoyouwe/Desktop/chord_data.json', 'w') as f:
    json.dump(data, f)

print("Data has been saved to 'chord_data.json'.")
```
